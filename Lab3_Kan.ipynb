### 2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
import pylab as plt
import numpy as np

a, b, c = 2, 1, 2 

x = np.linspace(-10, 10, 400)

y = a * x**2 + b * x + c

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = ax^2 + bx + c$')

plt.title('График квадратичной функции', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
```

### 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
import pylab as plt
import numpy as np

A = 2
omega = 2 * np.pi
phi = np.pi / 2

x = np.linspace(0, 2 * np.pi, 100)

y = A * np.sin(omega * x + phi)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = A \sin(\omega x + \phi)$')

plt.title('График гармонической функции', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
```

### 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
import pylab as plt
import numpy as np

x1 = np.linspace(-10, 10, 400)
y1 = a * x1**2 + b * x1 + c

x2 = np.linspace(0, 2 * np.pi, 100)
y2 = A * np.sin(omega * x2 + phi)

plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(x1, y1, label=r'$y = ax^2 + bx + c$')
plt.title('График квадратичной функции', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x2, y2, label=r'$y = A \sin(\omega x + \phi)$')
plt.title('График гармонической функции', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
import pylab as plt
import numpy as np

plt.figure(figsize=(8, 6))

plt.plot(x1, y1, label=r'$y = ax^2 + bx + c$', color='blue')

plt.plot(x2, y2, label=r'$y = A \sin(\omega x + \phi)$', color='red')

plt.title('Графики квадратичной и гармонической функций', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
```

### 6.	Построить график функции, заданной параметрически, x(t)=cos(3t), y=sin(2t), t ϵ [0,4π]

```python
t = np.linspace(0.4*np.pi, 300)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$x(t) = \cos(3t), y(t) = \sin(2t)$')

plt.title('Параметрический график функции', fontsize=14)
plt.xlabel('x(t)', fontsize=12)
plt.ylabel('y(t)', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
```

### 7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице
#### Наименование показателей 	2012	2013	2014	2015	2016	2017	2018	2019	2020	2021	2022
#### _________________________________________________________________________________________________________________
#### Кыргызская Республика	4341.15	4599.21	4981.51	5182.99	4794.34	4900.79	4792.54	4806.32	5358.53	6268.31	7178.32



```python
import matplotlib.pyplot as plt
import numpy as np

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
minimum_wage = [
    4341.15, 4599.21, 4981.51, 5182.99, 4794.34,
    4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(10, 6))
plt.bar(years, minimum_wage, color='skyblue')

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)', fontsize=16)
plt.xlabel('Год', fontsize=14)
plt.ylabel('Прожиточный минимум (сом)', fontsize=14)

for i, value in enumerate(minimum_wage):
    plt.text(years[i], value + 100, f'{value:.2f}', ha='center', va='bottom')

plt.xticks(years)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### 8.	Построить круговой график по данным п. 7

```python
import matplotlib.pyplot as plt

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
values = [
    4341.15, 4599.21, 4981.51, 5182.99,
    4794.34, 4900.79, 4792.54, 4806.32,
    5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(8, 8))
plt.pie(values, labels=years, autopct='%1.1f%%', startangle=140)

plt.title('Круговой график показателей Кыргызской Республики по годам')

plt.axis('equal')
plt.show()
```

### 9.	Построить график произвольной каркасной поверхности.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2, 3, 100)
y = np.linspace(-2, 3, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, color='blue', alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
